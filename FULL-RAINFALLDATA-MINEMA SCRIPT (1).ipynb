# 3. 
## AUTOMATIC RAINGAUGE MERGING AND AGGREGATION(SUMMING UP) 

```python
import pandas as pd

# =========================
# LOAD DATA
# =========================
rainfall_file = "/home/radar/Downloads/AWSRAINGAUGE-14.csv"
district_file = "/home/radar/Downloads/Raingauage_station_metadata.csv"   # file containing StationId and District

df = pd.read_csv(rainfall_file)
districts = pd.read_csv(district_file)

# =========================
# CLEAN COLUMN NAMES
# =========================
df.columns = df.columns.str.strip()
districts.columns = districts.columns.str.strip()

# =========================
# CONVERT RAINFALL TO NUMERIC
# =========================
df["AWSPRECIP"] = pd.to_numeric(df["AWSPRECIP"], errors="coerce").fillna(0)

# =========================
# FILTER DAYS 27 AND 28
# =========================
df_27_28 = df[df["Day"].isin([13, 14])].copy()

# =========================
# SUM RAINFALL PER STATION
# =========================
final_df = (
    df_27_28
    .groupby(
        ["StationId", "Station_Name", "Lat", "Lon", "Elev", "Year", "Month"],
        as_index=False
    )["AWSPRECIP"]
    .sum()
)

final_df = final_df.rename(columns={
    "AWSPRECIP": "Total_Rainfall_27_28"
})

# =========================
# MERGE DISTRICTS
# =========================
# District file should have at least:
# StationId, District

final_df = final_df.merge(
    districts[["StationId", "District"]],
    on="StationId",
    how="left"
)

# =========================
# REORDER COLUMNS
# =========================
final_df = final_df[
    [
        "StationId",
        "Station_Name",
        "District",
        "Lat",
        "Lon",
        "Elev",
        "Year",
        "Month",
        "Total_Rainfall_27_28"
    ]
]

# =========================
# SAVE OUTPUT
# =========================
final_df.to_csv("/home/radar/Downloads/13-Updated_Rainfall_Total_With_District.csv", index=False)

print(final_df)
print("File saved as: Updated_Rainfall_Total_With_District.xlsx")
```

## Removing brackets together numbers inside it 

```python
import pandas as pd
import re

# Corrected file path (use raw string r"..." or double backslashes \\ properly)
file_path = r"C:\\Users\\GAHIGI Robert\\OneDrive\\Documents\\Desktop\\MINEMA-AWS-2026.csv"

# Load your dataset
df = pd.read_csv(file_path)

# Function to remove numbers and brackets
def remove_numbers_and_brackets(text):
    if pd.isna(text):
        return text
    # First remove numbers
    text = re.sub(r'\d+', '', text)
    # Then remove brackets
    text = re.sub(r'[()]', '', text)
    return text.strip()

# Apply the function to the column you want to clean
df['StationName'] = df['StationName'].apply(remove_numbers_and_brackets)

# Save the cleaned dataset
df.to_csv('C:\\Users\\GAHIGI Robert\\OneDrive\\Desktop\\03_06_Removed_numbers_and_brackets_cleaned_dataset.csv', index=False)

print("Numbers and brackets removed successfully!")
```

## CLIMSOFT-DATABASE COMPATIBLE FORMAT converting script

```python
import pandas as pd
from datetime import datetime
import numpy as np

# Load the Mulindi CSV file
input_file = "C:\\Users\\GAHIGI Robert\\OneDrive\\Desktop\\PRACTICAL RAINFALL DATA\\MULINDI MIN&MAX_Temp(Feb-March).csv"
df = pd.read_csv(input_file)

# Parse the 'Recorded on' column as datetime with correct format (DD/MM/YYYY HH:MM)
df['Recorded on'] = pd.to_datetime(df['Recorded on'], format='%d/%m/%Y %H:%M', errors='coerce')

# Extract date parts
df['Year'] = df['Recorded on'].dt.year
df['Month'] = df['Recorded on'].dt.month
df['Day'] = df['Recorded on'].dt.day

# Define mapping of element names to Climsoft ElementIDs
element_map = {
    'Max Temp (oC)': 2,  # Max Temperature
    'Min Temp (oC)': 3,  # Min Temperature
   # 'Rainfall (mm)': 5   # Rainfall
}

# Prepare list for the transformed Climsoft-style records
climsoft_records = []

# Convert data into Climsoft format
for column, element_id in element_map.items():
    for _, row in df.iterrows():
        value = row[column]
        try:
            value = float(value)
        except (ValueError, TypeError):
            continue  # Skip rows with invalid data

        climsoft_records.append({
            'StationId': row['StationID'],
            'Year': row['Year'],
            'Month': row['Month'],
            'Day': row['Day'],
            'ElementID': element_id,
            'Value': value
        })

# Create DataFrame from records
climsoft_df = pd.DataFrame(climsoft_records)

# Save to CSV file
output_file = "C:\\Users\\GAHIGI Robert\\OneDrive\\Desktop\\updated_mulindi_climsoft_format.csv"
climsoft_df.to_csv(output_file, index=False)

print(f"✅ Climsoft-compatible file saved as: {output_file}")
```

```python
import pandas as pd

# === Step 1: Load the CSV file ===
file_path = "C:\\Users\\GAHIGI Robert\\OneDrive\\Desktop\\PRACTICAL RAINFALL DATA\\TAMIRA AWS.csv" # Change path if needed
df = pd.read_csv(file_path)

# === Step 2: Define your mapping ===
# Map combined "Measurement ID.Process ID" to human-readable names
id_mapping = {
    "1.1": "Temperature_Instant",
    "2.3": "Humidity_Minimum",
    "2.4": "Humidity_Maximum",
    "10.7":"Precipitation_Acummulation",
    "13.3":"Pressure_Minimum",
    "2.2":"Humidity_Average",
    # Add more mappings as needed
}

# === Step 3: Extract relevant columns based on mapping ===
new_columns = {}
i = 0
columns = df.columns.tolist()

while i < len(columns) - 2:
    try:
        # Read Measurement ID and Process ID from first row of each block
        meas_id = str(int(df.iloc[0, i]))
        proc_id = str(int(df.iloc[0, i + 1]))
        combined_key = f"{meas_id}.{proc_id}"

        # Check if this combination is in your mapping
        parameter_name = id_mapping.get(combined_key, None)
        if parameter_name:
            # Extract the data column (i+2), skipping the header row
            new_columns[parameter_name] = df.iloc[1:, i + 2].reset_index(drop=True)
        i += 3  # Move to the next block
    except:
        i += 1  # Skip over if anything unexpected happens

# === Step 4: Save the extracted DataFrame to CSV ===
result_df = pd.DataFrame(new_columns)
output_file = "C:\\Users\\GAHIGI Robert\\OneDrive\\Desktop\\processed_parameters.csv"
result_df.to_csv(output_file, index=False)

print(f"Processed data saved to '{output_file}'")
```
