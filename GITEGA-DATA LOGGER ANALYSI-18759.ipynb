```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
   
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
   
    outliers = data[(data[column] < lower) | (data[column] > upper)]
   
    return len(outliers), lower, upper
```

```python
def detect_outliers_variable_specific(df, col):
   
    data = df[col].dropna()
   
    # Default
    method = "IQR"
    lower, upper = None, None
   
    col_lower = col.lower()
   
    # =========================
    # 1. Temperature
    # =========================
    if "temp" in col_lower:
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        method = "IQR (Temperature)"
   
    # =========================
    # 2. Humidity
    # =========================
    elif "humid" in col_lower:
        lower = 0
        upper = 100
        method = "Physical Range (0–100%)"
   
    # =========================
    # 3. Precipitation
    # =========================
    # elif "rain" in col_lower or "precip" in col_lower:
        #lower = 0
        #upper = data.quantile(0.99)  # extreme rainfall threshold
        #method = "99th Percentile (Precipitation)"
    #if "Precip" in col_lower:
        #Q1 = data.quantile(0.25)
        #Q3 = data.quantile(0.75)
        #IQR = Q3 - Q1
        #lower = Q1 - 1.5 * IQR
        #upper = Q3 + 1.5 * IQR
        #method = "IQR (precipitation)"
    elif "Precip" in col_lower:
        lower = 0
        upper = 100
        method = "Physical Range (0–200mm)"
   
    # =========================
    # 4. Wind
    # =========================
    elif "wind" in col_lower:
        lower = 0
        upper = data.quantile(0.99)
        method = "Range + Percentile (Wind)"
   
    # =========================
    # 5. Default (fallback)
    # =========================
    else:
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        method = "IQR (Default)"
   
    # Detect outliers
    outliers = df[(df[col] < lower) | (df[col] > upper)]
   
    return len(outliers), lower, upper, method
```

```python
outlier_results = []

for col in numeric_cols:
    count, lower, upper, method = detect_outliers_variable_specific(df, col)
   
    outlier_results.append({
        "Variable": col,
        "Outliers_Count": count,
        "Lower_Bound": lower,
        "Upper_Bound": upper,
        "Method_Used": method
    })

outlier_df = pd.DataFrame(outlier_results)
display(outlier_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# -----------------------------
# Sort by date
# -----------------------------
df = df.sort_values("Date").copy()

# -----------------------------
# Homogeneity check function
# -----------------------------
def check_homogeneity(data, column, mean_threshold=0.2, std_threshold=0.3, min_obs=20):
    s = data[column].dropna()
   
    if len(s) < min_obs:
        return {
            "Variable": column,
            "N": len(s),
            "Mean_First": np.nan,
            "Mean_Second": np.nan,
            "Std_First": np.nan,
            "Std_Second": np.nan,
            "Mean_Change_Ratio": np.nan,
            "Std_Change_Ratio": np.nan,
            "Flag": "Insufficient Data"
        }
   
    mid = len(s) // 2
    first = s.iloc[:mid]
    second = s.iloc[mid:]
   
    mean_first = first.mean()
    mean_second = second.mean()
    std_first = first.std()
    std_second = second.std()
   
    mean_base = abs(mean_first) if abs(mean_first) > 1e-6 else 1
    std_base = abs(std_first) if abs(std_first) > 1e-6 else 1
   
    mean_change_ratio = abs(mean_second - mean_first) / mean_base
    std_change_ratio = abs(std_second - std_first) / std_base
   
    if (mean_change_ratio > mean_threshold) or (std_change_ratio > std_threshold):
        flag = "Potential Issue"
    else:
        flag = "Stable"
   
    return {
        "Variable": column,
        "N": len(s),
        "Mean_First": mean_first,
        "Mean_Second": mean_second,
        "Std_First": std_first,
        "Std_Second": std_second,
        "Mean_Change_Ratio": mean_change_ratio,
        "Std_Change_Ratio": std_change_ratio,
        "Flag": flag
    }

# -----------------------------
# Apply check to all variables
# -----------------------------
results = [check_homogeneity(df, col) for col in analysis_cols]
homogeneity_df = pd.DataFrame(results)

# -----------------------------
# Severity classification
# -----------------------------
def severity_level(row):
    if row["Flag"] == "Insufficient Data":
        return "Insufficient Data"
   
    score = max(row["Mean_Change_Ratio"], row["Std_Change_Ratio"])
   
    if score < 0.2:
        return "Low"
    elif score < 0.5:
        return "Moderate"
    else:
        return "High"

homogeneity_df["Severity"] = homogeneity_df.apply(severity_level, axis=1)

# -----------------------------
# Round detailed table
# -----------------------------
homogeneity_df = homogeneity_df.round(3)

print("=" * 80)
print("DETAILED HOMOGENEITY RESULTS")
print("=" * 80)
print(homogeneity_df)

# -----------------------------
# One-row summary table
# -----------------------------
summary_row = {
    "Total_Variables": len(homogeneity_df),
    "Stable": (homogeneity_df["Flag"] == "Stable").sum(),
    "Potential_Issues": (homogeneity_df["Flag"] == "Potential Issue").sum(),
    "Insufficient_Data": (homogeneity_df["Flag"] == "Insufficient Data").sum(),
    "Low_Severity": (homogeneity_df["Severity"] == "Low").sum(),
    "Moderate_Severity": (homogeneity_df["Severity"] == "Moderate").sum(),
    "High_Severity": (homogeneity_df["Severity"] == "High").sum()
}

summary_df = pd.DataFrame([summary_row])

print("\n" + "=" * 80)
print("HOMOGENEITY SUMMARY (ONE ROW)")
print("=" * 80)
print(summary_df)

# -----------------------------
# Table of only flagged variables
# -----------------------------
flagged_df = homogeneity_df[homogeneity_df["Flag"] == "Potential Issue"].copy()

print("\n" + "=" * 80)
print("FLAGGED VARIABLES ONLY")
print("=" * 80)
print(flagged_df)

# -----------------------------
# Optional: one-row wide table of flags
# -----------------------------
flag_row_df = homogeneity_df.set_index("Variable")["Flag"].to_frame().T
flag_row_df.index = ["Homogeneity_Status"]

print("\n" + "=" * 80)
print("ONE-ROW STATUS TABLE")
print("=" * 80)
print(flag_row_df)

# -----------------------------
# Plot flagged variables
# -----------------------------
flagged_vars = flagged_df["Variable"].tolist()

if len(flagged_vars) > 0:
    cols = 2
    rows = math.ceil(len(flagged_vars) / cols)
   
    plt.figure(figsize=(14, rows * 3))
   
    for i, col in enumerate(flagged_vars):
        plt.subplot(rows, cols, i + 1)
       
        rolling = df[col].rolling(window=30).mean()
       
        plt.plot(df["Date"], df[col], alpha=0.4, linewidth=1, label="Original")
        plt.plot(df["Date"], rolling, linewidth=2, label="Rolling Mean")
       
        plt.title(col, fontsize=10)
        plt.xlabel("")
        plt.ylabel("")
        plt.grid(True, alpha=0.3)
        plt.legend(fontsize=8)
   
    plt.suptitle("Variables Flagged for Potential Homogeneity Issues", fontsize=14, weight="bold")
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
else:
    print("\nNo variables were flagged for homogeneity issues.")
```

# Outlier detection using Z-score 

```python
from scipy.stats import zscore
import numpy as np

def detect_outliers_zscore(data, column, threshold=3):
    clean_data = data[column].dropna()
   
    z_scores = np.abs(zscore(clean_data))
   
    outlier_index = clean_data.index[z_scores > threshold]
   
    return data.loc[outlier_index]
```

```python
zscore_results = []

for col in analysis_cols:
    outliers = detect_outliers_zscore(df, col, threshold=3)
   
    zscore_results.append({
        "Variable": col,
        "Outliers": len(outliers)
    })

zscore_df = pd.DataFrame(zscore_results)
print(zscore_df)
```

## Task 2: Calculation of climate indices 
### step 1: Prepare the dataset  and define the consecutive days function 

```python
import pandas as pd
import numpy as np

# Ensure correct types
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df["Precip"] = pd.to_numeric(df["Precip"], errors="coerce")

# Build daily dataframe with sum + count
df_daily = (
    df.set_index("Date")
      .resample("D")
      .agg(
          Daily_Precip=("Precip", "sum"),
          Obs_Count=("Precip", "count")
      )
      .reset_index()
)

# Extract year
df_daily["Year"] = df_daily["Date"].dt.year

# Mark days with no observations as missing
df_daily.loc[df_daily["Obs_Count"] == 0, "Daily_Precip"] = np.nan

print(df_daily.head())
```

### step 2: CDD and CWD 

```python
def max_consecutive_days(condition_series):
    max_count = 0
    count = 0

    for val in condition_series:
        if val:
            count += 1
            max_count = max(max_count, count)
        else:
            count = 0

    return max_count

# CDD: daily rainfall < 1 mm
cdd = df_daily.groupby("Year")["Daily_Precip"].apply(
    lambda x: max_consecutive_days(x < 1)
)

# CWD: daily rainfall >= 1 mm
cwd = df_daily.groupby("Year")["Daily_Precip"].apply(
    lambda x: max_consecutive_days(x >= 1)
)

print("CDD (Consecutive Dry Days):")
print(cdd)

print("\nCWD (Consecutive Wet Days):")
print(cwd)
```

## TASK 4: SEASONAL COMPARISON 

### 1. We first define the seasons of Rwanda 

```python
def get_season(month):
    if month in [1, 2]:
        return "JF"
    elif month in [3, 4, 5]:
        return "MAM"
    elif month in [6, 7, 8]:
        return "JJA"
    else:
        return "SOND"

df_daily_clean["Month"] = df_daily_clean["Date"].dt.month
df_daily_clean["Season"] = df_daily_clean["Month"].apply(get_season)
```

### step 2: Compute the seasonal rainfall

```python
seasonal_rain = df_daily_clean.groupby(["Year", "Season"])["Daily_Precip"].sum().reset_index()

print(seasonal_rain.head(10))
```

### step 3: Pivot the data for clarity

```python
seasonal_table = seasonal_rain.pivot(index="Year", columns="Season", values="Daily_Precip")

print(seasonal_table)
```

### seasonal Rainfall trends 

```python
from scipy.stats import linregress

season_trends = []

for season in seasonal_table.columns:
    s = seasonal_table[season].dropna()
   
    if len(s) >= 2:
        x = s.index.values
        y = s.values
       
        slope, _, _, p_value, _ = linregress(x, y)
       
        season_trends.append({
            "Season": season,
            "Trend_Slope": slope,
            "p_value": p_value
        })

season_trends_df = pd.DataFrame(season_trends)

print(season_trends_df)
```

## 🧠 Interpretation

slope > 0 → increasing rainfall

slope < 0 → decreasing

p < 0.05 → significant
